**Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.**

Solution:

Matplotlib is a popular data visualization library in Python that provides a wide range of functionalities for creating various types of plots and charts. It is widely used for creating static, animated, and interactive visualizations in scientific computing, data analysis, and other fields.

Matplotlib is used for visualizing data to gain insights, present findings, and communicate information effectively. It offers a high level of customization, allowing users to create visually appealing and informative plots.

The Pyplot module within Matplotlib provides a simple and convenient interface for creating plots. It offers a collection of functions that allow users to create different types of plots. Here are five common types of plots that can be created using the Pyplot module of Matplotlib:

1. Line Plot: A line plot represents data points connected by straight lines. It is commonly used to show trends and patterns over time or any ordered sequence.

2. Bar Plot: A bar plot displays rectangular bars of varying heights to represent data. It is often used to compare values across different categories or to show the distribution of a single variable.

3. Scatter Plot: A scatter plot shows the relationship between two variables by displaying individual data points as dots on a two-dimensional plane. It helps visualize patterns, correlations, or clusters in the data.

4. Histogram: A histogram represents the distribution of a continuous variable by dividing the data into bins and displaying the frequency or count of values falling into each bin. It provides insights into the underlying distribution of the data.

5. Pie Chart: A pie chart is a circular plot divided into sectors that represent different categories or proportions of a whole. It is used to display the composition or distribution of categorical data.

These are just a few examples of the types of plots that can be created using the Pyplot module of Matplotlib. Matplotlib offers many more plot types and customization options to cater to a wide range of data visualization needs.

**Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.**

**import numpy as np**

**np.random.seed(3)**

**x = 3 + np.random.normal(0, 2, 50)**

**y = 3 + np.random.normal(0, 2, len(x))**

**Note: Also add title, xlabel, and ylabel to the plot.**
    
Solution:

A scatter plot is a type of plot used to visualize the relationship between two continuous variables. It represents individual data points as dots on a two-dimensional plane, with one variable plotted on the x-axis and the other variable plotted on the y-axis. Scatter plots are useful for identifying patterns, correlations, or clusters in the data.

To generate a scatter plot using the provided code, we can use the Matplotlib library, specifically the Pyplot module. 

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create scatter plot
plt.scatter(x, y)

# Add title, xlabel, and ylabel
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')

# Display the plot
plt.show()
```

**Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])**

Solution:

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple smaller plots and plot different data or visualizations in each subplot. The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current subplot.

Using the subplot() function, you can create multiple line plots within a single figure, each occupying its own subplot. Here's the code to draw four line plots using the subplot() function with the provided data:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

# Adjust layout to prevent overlapping of subplots
plt.tight_layout()

# Display the plot
plt.show()
```

**Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])**

Solution:
    
A bar plot, also known as a bar chart or bar graph, is a type of plot that represents categorical data using rectangular bars of varying heights. The length of each bar corresponds to the value of the data it represents. Bar plots are commonly used to compare values across different categories or to display the distribution of a single variable.

Bar plots are used for various purposes, including:

Comparison: Bar plots allow for easy comparison of values between different categories. They visually highlight the differences or similarities in values, making it useful for analyzing data and identifying patterns or trends.

Ranking: Bar plots can be used to rank categories based on their values. This ranking can help identify the top or bottom performers, such as companies with the highest or lowest profits.

Distribution: Bar plots can display the distribution of a single variable across different categories. This can help identify patterns or outliers within the data.


using the provided data to create a bar plot and a horizontal bar plot using the Matplotlib library:

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title('Company Profit')
plt.xlabel('Company')
plt.ylabel('Profit')

plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title('Company Profit')
plt.xlabel('Profit')
plt.ylabel('Company')

plt.show()
```

**Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)**

Solution:

A box plot, also known as a box-and-whisker plot, is a type of plot used to display the distribution of a dataset. It provides a visual summary of key statistical measures such as the median, quartiles, and potential outliers. A box plot consists of a rectangle (the "box") and two lines (the "whiskers") that extend from the box, along with individual data points (if any) beyond the whiskers.

Box plots are commonly used for the following purposes:

Summary of Distribution: A box plot provides a compact summary of the distribution of a dataset, allowing for quick assessment of its central tendency, spread, skewness, and potential outliers.

Comparison: Box plots enable easy visual comparison of multiple datasets or groups. By placing multiple box plots side by side, one can compare their distributions and identify differences in measures such as medians, quartiles, or ranges.

Outlier Detection: Box plots help identify potential outliers in a dataset. Outliers are represented as individual data points beyond the whiskers, indicating values that are significantly different from the majority of the data.

Now, let's create a box plot using the provided data using the Matplotlib library:

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Value')

plt.show()
```
