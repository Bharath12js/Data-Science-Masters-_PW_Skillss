# SOLUTIONS:

```python
import numpy as np
import scipy.stats as stats

# Q1: Calculate the 95% confidence interval
sample_mean = 50
sample_std = 5
sample_size = 100  # You need to provide the sample size

confidence_level = 0.95
margin_of_error = stats.norm.ppf(1 - (1 - confidence_level) / 2) * (sample_std / np.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("Q1: 95% Confidence Interval:", confidence_interval)

# Q2: Chi-square goodness of fit test
observed_frequencies = np.array([25, 30, 25, 15, 10, 20])  # Provide observed frequencies
expected_frequencies = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed_frequencies)

chi_square, p_value = stats.chisquare(observed_frequencies, f_exp=expected_frequencies)

print("Q2: Chi-square statistic:", chi_square)
print("P-value:", p_value)
if p_value < 0.05:
    print("Reject the null hypothesis. The distribution does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The distribution matches the expected distribution.")

# Q3: Chi-square test for a contingency table
observed_data = np.array([[20, 15], [10, 25], [15, 20]])  # Provide observed data

chi2, p, dof, expected = stats.chi2_contingency(observed_data)

print("Q3: Chi-square statistic:", chi2)
print("P-value:", p)
if p < 0.05:
    print("Reject the null hypothesis. There is a significant association.")
else:
    print("Fail to reject the null hypothesis. There is no significant association.")

# Q4: Calculate the 95% confidence interval for proportion
sample_size = 500
observed_smokers = 60

proportion = observed_smokers / sample_size

conf_int = stats.norm.interval(0.95, loc=proportion, scale=np.sqrt((proportion * (1 - proportion)) / sample_size))

print("Q4: 95% Confidence Interval:", conf_int)

# Q5: Calculate the 90% confidence interval
sample_mean = 75
sample_std = 12
sample_size = 50  # You need to provide the sample size

confidence_level = 0.90
margin_of_error = stats.norm.ppf(1 - (1 - confidence_level) / 2) * (sample_std / np.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("Q5: 90% Confidence Interval:", confidence_interval)

# Q6: Plot the chi-square distribution
import matplotlib.pyplot as plt

df = 10
x = np.linspace(0, 30, 1000)
y = stats.chi2.pdf(x, df)

plt.plot(x, y, label=f"Chi-square Distribution (df = {df})")
plt.axvline(x=15, color='r', linestyle='--', label="Chi-square Statistic = 15")
plt.xlabel('Chi-square Value')
plt.ylabel('Probability Density')
plt.title('Chi-square Distribution')
plt.legend()
plt.show()

# Q7: Calculate a 99% confidence interval for proportion
sample_size = 1000
observed_coke_preferred = 520

proportion = observed_coke_preferred / sample_size

conf_int = stats.norm.interval(0.99, loc=proportion, scale=np.sqrt((proportion * (1 - proportion)) / sample_size))

print("Q7: 99% Confidence Interval:", conf_int)

# Q8: Chi-square goodness of fit test for coin flip
observed_frequencies = np.array([45, 55])  # Provide observed frequencies
expected_frequencies = np.array([0.5, 0.5]) * np.sum(observed_frequencies)

chi_square, p_value = stats.chisquare(observed_frequencies, f_exp=expected_frequencies)

print("Q8: Chi-square statistic:", chi_square)
print("P-value:", p_value)
if p_value < 0.05:
    print("Reject the null hypothesis. The observed frequencies do not match the expected frequencies of a fair coin.")
else:
    print("Fail to reject the null hypothesis. The observed frequencies match the expected frequencies of a fair coin.")

# Q9: Chi-square test for independence
observed_data = np.array([[20, 15], [10, 25], [15, 20], [60, 140], [30, 170]])  # Provide observed data

chi2, p, dof, expected = stats.chi2_contingency(observed_data)

print("Q9: Chi-square statistic:", chi2)
print("P-value:", p)
if p < 0.05:
    print("Reject the null hypothesis. There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between smoking status and lung cancer diagnosis.")

# Q10: Chi-square test for independence (additional question)
observed_data = np.array([[120, 220, 160], [150, 180, 170]])  # Provide observed data

chi2, p, dof, expected = stats.chi2_contingency(observed_data)

print("Q10: Chi-square statistic:", chi2)
print("P-value:", p)
if p < 0.01:
    print("Reject the null hypothesis. There is a significant association between chocolate preference and country of origin.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between chocolate preference and country of origin.")

# Q11: Hypothesis test for population mean
sample_mean = 72
sample_std = 10
sample_size = 30
population_mean = 70

alpha = 0.05

t_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=sample_size - 1))

print("Q11: t-statistic:", t_statistic)
print("P-value:", p_value)
if p_value < alpha:
    print("Reject the null hypothesis. The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in population mean from 70.")
```
