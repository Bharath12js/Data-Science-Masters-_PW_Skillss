Q1: The Probability Mass Function (PMF) and Probability Density Function (PDF) are mathematical functions used to describe the probability distribution of a random variable.

- PMF: The PMF is used for discrete random variables and gives the probability of each possible outcome. It maps each value of the random variable to its corresponding probability. The sum of all probabilities in the PMF equals 1. An example is the fair six-sided die, where the PMF would assign a probability of 1/6 to each possible outcome (1, 2, 3, 4, 5, 6).

- PDF: The PDF is used for continuous random variables and represents the relative likelihood of the random variable taking on a specific value. Unlike the PMF, the PDF does not give the probability of a specific outcome, but rather the probability density at a particular point. The probability of an event occurring within a range is calculated by integrating the PDF over that range. For example, the PDF of the standard normal distribution (mean = 0, standard deviation = 1) describes the shape of the bell curve and provides the likelihood of observing a particular value.

Q2: The Cumulative Distribution Function (CDF) gives the probability that a random variable takes on a value less than or equal to a given value. It provides information about the probability distribution of a random variable over its entire range.

The CDF is defined for both discrete and continuous random variables. For discrete random variables, the CDF is the cumulative sum of the probabilities from the PMF up to a given value. For continuous random variables, it is obtained by integrating the PDF from negative infinity up to a given value.

The CDF is used to calculate probabilities for specific ranges or intervals and to determine percentiles. It can also be used to assess the probability of observing values within a certain range.

For example, the CDF of a standard normal distribution tells us the probability that a randomly chosen observation from the distribution will be less than or equal to a specific value.

Q3: The normal distribution is commonly used as a model in situations where data clusters around a central value with fewer extreme values on either side. Some examples where the normal distribution might be used as a model include:

- Heights of a population: The heights of adult males or females often follow a roughly normal distribution, with most individuals clustering around the mean height and fewer individuals at the extreme ends (very short or very tall).

- Errors in measurements: Errors in measurements, such as in scientific experiments or industrial processes, are often assumed to be normally distributed. The mean represents the expected value, and the standard deviation represents the variability or precision of the measurements.

- IQ scores: IQ scores are often modeled using a normal distribution. The mean represents the average IQ score, and the standard deviation represents the spread or variation in IQ scores.

The parameters of the normal distribution, namely the mean (μ) and standard deviation (σ), determine the shape of the distribution. The mean specifies the central location of the distribution, while the standard deviation controls the spread or width of the distribution. Larger standard deviations result in broader distributions, while smaller standard deviations result in narrower distributions.

- It is a widely used model for many natural and social phenomena. Numerous real-life measurements, such as heights, weights, test scores, and IQ scores, tend to follow a normal distribution.

- The Central Limit Theorem states that the sum or average of a large number of independent and identically distributed random variables will be approximately normally distributed, regardless of the shape of the original distribution. - The normal distribution has well-defined properties and mathematical properties, making it amenable to analysis and calculations. Many statistical methods and hypothesis tests are built on the assumptions of normality.

- Heights of adult males or females: Heights in a population often exhibit a normal distribution, with the mean height being around the average height of the population.

It has a probability parameter, usually denoted as "p," which represents the probability of success. An example could be flipping a fair coin, where success (1) represents heads and failure (0) represents tails.

The difference between Bernoulli Distribution and Binomial Distribution lies in the number of trials. Bernoulli Distribution describes a single trial, whereas Binomial Distribution describes the number of successes in a fixed number of independent Bernoulli trials. Binomial Distribution extends the concept of Bernoulli Distribution to multiple trials.

Q6: To find the probability that a randomly selected observation from a normally distributed dataset with a mean of 50 and a standard deviation of 10 is greater than 60, we can use the Standard Normal Distribution (Z-distribution).

First, we standardize the value of 60 using the formula:

Z = (x - μ) / σ

Where:
x = 60 (the value we want to calculate the probability for)
μ = 50 (mean of the dataset)
σ = 10 (standard deviation of the dataset)

Z = (60 - 50) / 10 = 1

Next, we look up the corresponding probability from the standard normal distribution table or use a calculator. The area to the right of Z = 1 represents the probability of a randomly selected observation being greater than 60.

Using the standard normal distribution table or a calculator, we find that the probability associated with Z = 1 is approximately 0.8413. Therefore, the probability that a randomly selected observation from the given dataset is greater than 60 is approximately 0.8413 or 84.13%.

Q7: The uniform distribution is a continuous probability distribution where all values within a specific range are equally likely to occur. It is characterized by a constant probability density function (PDF) over its support interval.

An example of a uniform distribution is rolling a fair six-sided die. Each outcome (1, 2, 3, 4, 5, 6) has an equal probability of 1/6. The PDF of a uniform distribution is a horizontal line over the range of possible values, indicating the constant probability.

The uniform distribution is often used when there is no reason to favor any specific outcome within a given range. It is also used as a building block for generating random numbers.

Q8: The z-score (also known as the standard score) is a measure of how many standard deviations an observation or data point is from the mean of a distribution. It is calculated using the formula:

z = (x - μ) / σ

Where:
x = the data point or observation
μ = the mean of the distribution
σ = the standard deviation of the distribution

The z-score allows us to compare and standardize data points from different distributions. It tells us how many standard deviations a data point is above or below the mean. A positive z-score indicates that the data point is above the mean, while a negative z-score indicates that it is below the mean.

The importance of the z

-score lies in its usefulness for hypothesis testing, statistical inference, and data analysis. It enables us to make comparisons, determine the relative position of a data point within a distribution, and calculate probabilities associated with specific values or ranges.

Q9: The Central Limit Theorem (CLT) states that the distribution of the sample means (or sums) of a large number of independent and identically distributed random variables approaches a normal distribution, regardless of the shape of the original population distribution. The CLT holds under certain conditions, including the requirement that the sample size is sufficiently large.

The significance of the Central Limit Theorem is that it allows us to make inferences about a population based on a sample, even when the population distribution is unknown or non-normal. It forms the basis for many statistical techniques and hypothesis tests, such as confidence intervals and hypothesis testing about means.

The CLT enables us to use the normal distribution as an approximation for the sampling distribution of the sample mean. This simplifies statistical analysis and allows for the application of various statistical tests that rely on the properties of the normal distribution.

Q10: The assumptions of the Central Limit Theorem include:

1. Independence: The random variables in the sample should be independent of each other. This means that the value of one observation does not influence the value of another observation.

2. Identically Distributed: The random variables should be drawn from the same population or have the same distribution. This assumption ensures that the observations have a similar underlying distribution.

3. Finite Variance: The random variables should have a finite variance. If the variance is infinite or does not exist, the CLT may not hold.

These assumptions, particularly the independence assumption, are crucial for the validity and applicability of the Central Limit Theorem. Violations of these assumptions can affect the accuracy of the theorem and may require alternative statistical methods.
