Q1. The three measures of central tendency are the mean, median, and mode.

Q2. The mean is the average of a dataset and is calculated by summing all the values and dividing by the number of observations. The median is the middle value in a dataset when it is arranged in ascending or descending order. The mode is the value that appears most frequently in a dataset.

These measures are used to determine the central tendency of a dataset, which is a way to describe the typical or central value around which the data points tend to cluster. The mean provides a measure of the average value, the median represents the middle value, and the mode identifies the most common value in the dataset.

μ = (178 + 177 + 176 + 177 + 178.2 + 178 + 175 + 179 + 180 + 175 + 178.9 + 176.2 + 177 + 172.5 + 178 + 176.5) / 16
   = 2816.3 / 16
   = 176.01875

(178 - 176.01875)^2, (177 - 176.01875)^2, (176 - 176.01875)^2, ...

Sum = (178 - 176.01875)^2 + (177 - 176.01875)^2 + (176 - 176.01875)^2 + ...

Finally, take the square root of the result to obtain the standard deviation.

Q5. Measures of dispersion such as range, variance, and standard deviation are used to describe the spread or variability of a dataset.

- Range: It measures the difference between the maximum and minimum values in a dataset. A larger range indicates a greater spread.

- Variance: It measures the average squared deviation from the mean. A larger variance indicates greater variability in the data.

- Standard Deviation: It is the square root of the variance. It measures the average deviation from the mean and provides a standardized measure of dispersion. A larger standard deviation indicates greater variability in the data.

For example, let's consider two datasets:
Dataset A: [10, 20, 30, 40, 50]
Dataset B: [10, 10, 10, 40, 70]

Both datasets have the same mean (30) but different measures of dispersion. Dataset B has a larger range, variance, and standard deviation compared to Dataset A, indicating a greater spread and variability in the data.

Q8. Skewness in data refers to the asymmetry in its distribution. It indicates whether the data is skewed or distorted towards one tail of the distribution. Skewness can be classified into three types: positive skewness (right-skewed), negative skewness (left-skewed), and zero skewness (symmetrical).

- Positive skewness (right-skewed): The distribution has a long right tail, with more data points towards the left. The mean is usually greater than the median.

- Negative skewness (left-skewed): The distribution has a long left tail, with more data points towards the right. The mean is usually smaller than the median.

- Zero skewness (symmetrical): The distribution is symmetrical, with equal amounts of data points on both sides. The mean and median are approximately equal.

Q9. If a data is right skewed, the median will be positioned to the left of the mean. In right-skewed distributions, the long tail on the right side of the distribution pulls the mean towards higher values, making it larger than the median.

Q10. Covariance and correlation are both measures of the relationship between two variables in statistical analysis, but they have distinct differences:

- Covariance: Covariance measures how two variables change together. It indicates the direction and strength of the linear relationship between variables. However, covariance does not provide a standardized measure, making it difficult to interpret the strength of the relationship. Covariance can take any value, positive or negative.

- Correlation: Correlation also measures the relationship between variables but provides a standardized measure between -1 and 1, which allows for easier interpretation. A correlation of 1 indicates a perfect positive relationship, -1 indicates a perfect negative relationship, and 0 indicates no linear relationship.

Covariance and correlation are used to determine the strength and

 direction of the relationship between variables. They are commonly used in various fields, such as finance, social sciences, and data analysis, to understand the associations between variables.

Q11. The formula for calculating the sample mean (x̄) is:
x̄ = (x₁ + x₂ + x₃ + ... + xₙ) / n

Example: Consider the dataset [10, 15, 20, 25, 30]. To calculate the sample mean:
x̄ = (10 + 15 + 20 + 25 + 30) / 5
   = 100 / 5
   = 20

Therefore, the sample mean of the given dataset is 20.

Q12. In a normal distribution, the measures of central tendency (mean, median, and mode) are equal. This is because a normal distribution is symmetric, with the highest frequency occurring at the mean. The mean is located at the center of the distribution, and the median also coincides with the mean. Additionally, since the distribution is symmetric, there is only one mode, which is equal to the mean and median.

Q13. Covariance and correlation are related but different measures:

- Covariance measures the direction and strength of the linear relationship between two variables. It can take any value, positive or negative, and is affected by the units of measurement of the variables. Covariance does not provide a standardized measure, making it difficult to interpret the strength of the relationship.

- Correlation measures the direction and strength of the linear relationship between two variables, but it provides a standardized measure between -1 and 1. Correlation is unaffected by the units of measurement, making it easier to compare the strength of relationships. A correlation of 1 indicates a perfect positive relationship, -1 indicates a perfect negative relationship, and 0 indicates no linear relationship.

In summary, covariance measures the relationship between variables in their original units, while correlation provides a standardized measure that allows for easier interpretation and comparison.

Q14. Outliers can significantly impact measures of central tendency and dispersion:

- Measures of Central Tendency: Outliers, which are extreme values that deviate significantly from the rest of the data, can heavily influence the mean, pulling it towards their value. The median is more resistant to outliers as it represents the middle value, but extreme outliers can still affect it. The mode is not affected by outliers.

- Measures of Dispersion: Outliers can increase the range, as they contribute to the difference between the maximum and minimum values. Outliers can also affect the variance and standard deviation by increasing their values, as they introduce greater variability into the dataset.

For example, consider the dataset [10, 15, 20, 25, 1000]. The mean is heavily influenced by the outlier (1000), resulting in a larger value compared to the other data points. The range is also affected, as the difference between the maximum (1000) and minimum (10) values becomes significantly larger. The outlier increases the variability in the dataset, leading to larger variance and standard deviation values.
