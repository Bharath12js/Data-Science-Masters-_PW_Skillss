Q1: The Difference between a t-test and a z-test:
A t-test and a z-test are both statistical hypothesis tests used to make inferences about population parameters based on sample data. The main difference between them is in the circumstances under which they are used, specifically when the population standard deviation is known (z-test) or unknown (t-test).

Example scenario for using a t-test:
Suppose you want to determine if there is a significant difference in the average scores of two groups of students (Group A and Group B) on a standardized test. However, you don't know the population standard deviation. In this case, you would use a t-test to compare the means of the two groups and assess whether the difference is statistically significant.

Example scenario for using a z-test:
Imagine you are conducting a study to compare the heights of adult males in a certain city to the national average height for adult males, and you have access to the population standard deviation of heights in that city. You could use a z-test to determine if the average height in the city is significantly different from the national average.

Q2: One-Tailed vs. Two-Tailed Tests:
In hypothesis testing, the distinction between one-tailed and two-tailed tests relates to the directionality of the hypothesis being tested and the corresponding critical region.

One-tailed test:
A one-tailed test is used when the research hypothesis specifies a particular direction of effect or difference. For instance, you might be testing whether a new drug increases reaction times. If you expect the drug to slow down reactions, you would use a one-tailed test. The critical region is only on one side of the distribution.

Two-tailed test:
A two-tailed test is appropriate when the research hypothesis is non-directional, meaning you are interested in any significant difference or effect, regardless of direction. For example, if you're investigating whether a coin is biased, you're interested in deviations from expected values in either direction (heads or tails). The critical region is split between both tails of the distribution.

Q3: Type 1 and Type 2 Errors:
Type 1 Error (False Positive): This occurs when you reject a true null hypothesis. In other words, you conclude there is an effect or difference when there actually isn't. It's akin to a "false alarm."

Example: A medical test incorrectly diagnoses a healthy patient as having a disease.

Type 2 Error (False Negative): This happens when you fail to reject a false null hypothesis. In this case, there is a real effect or difference, but your test fails to detect it. It's like missing an actual event.

Example: A new drug has a therapeutic effect, but a clinical trial fails to detect this effect, leading to a wrong conclusion.

Q4: Bayes's Theorem:
Bayes's Theorem is a mathematical formula used to update probabilities based on new evidence. It's used in probability and statistics to revise beliefs or probabilities as new information becomes available.

The formula: P(A|B) = (P(B|A) * P(A)) / P(B)

Where:
- P(A|B) is the probability of event A occurring given that event B has occurred.
- P(B|A) is the probability of event B occurring given that event A has occurred.
- P(A) is the prior probability of event A.
- P(B) is the prior probability of event B.

Example: Medical Test
Suppose you're testing for a rare disease, and the test is 99% accurate. The disease affects 0.1% of the population. If the test is positive (B), what is the probability that the person actually has the disease (A)?

- P(A) = 0.001 (0.1%)
- P(B|A) = 0.99 (Test is positive given the person has the disease)
- P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)
- P(B|not A) = 1 - P(B|A) = 0.01 (Test is positive given the person doesn't have the disease)
- P(not A) = 1 - P(A) = 0.999

Now plug into Bayes's Theorem:
P(A|B) = (P(B|A) * P(A)) / P(B)
P(A|B) = (0.99 * 0.001) / [(0.99 * 0.001) + (0.01 * 0.999)]
P(A|B) ≈ 0.09 or 9%

Even with a positive test result, the probability of having the disease is still only around 9% due to the low base rate.

Q5: Confidence Interval and Calculation:
A confidence interval is a range of values within which we estimate a population parameter (like a mean or proportion) to lie, with a certain level of confidence.

For a population mean with a known standard deviation:
Confidence Interval = (Sample Mean) ± (Z * (Standard Deviation / √n))

Where:
- Z is the critical value based on the desired confidence level (e.g., 1.96 for a 95% confidence level).
- Standard Deviation is the population standard deviation.
- n is the sample size.

Example:
You have a sample of exam scores with a mean of 75, a known population standard deviation of 10, and a sample size of 100. Calculate a 95% confidence interval for the population mean.

Confidence Interval = 75 ± (1.96 * (10 / √100))
Confidence Interval = 75 ± 1.96

Interpretation:
We are 95% confident that the true population mean lies within the interval [73.04, 76.96].

Q6: Bayes's Theorem with New Evidence:
Let's say you are testing a new diagnostic tool for a medical condition. Based on previous studies, you know that the tool correctly identifies the condition 98% of the time (P(B|A)), and the condition occurs in 2% of the population (P(A)). If the tool tests positive (B), what is the probability that the person actually has the condition (A)?

- P(A) = 0.02
- P(B|A) = 0.98
- P(B|not A) = 1 - P(B|A) = 0.02
- P(not A) = 1 - P(A) = 0.98

Using Bayes's Theorem:
P(A|B) = (P(B|A) * P(A)) / P(B)
P(A|B) = (0.98 * 0.02) / [(0.98 * 0.02) + (0.02 * 0.98)]
P(A|B) = 0.5 or 50%

Even with a positive test result, the probability of having the condition is only 50% due to the relatively low prevalence of the condition in the population.

Q7: Calculation of a 95% Confidence Interval:
Given a sample mean (x̄) of 50 and a standard deviation (s) of 5, the formula for a 95% confidence interval for the population mean (μ) is:

Confidence Interval = x̄ ± (Z * (s / √n))

 Where:
-

 Z is the critical value for a 95% confidence level, approximately 1.96.
- s is the sample standard deviation.
- n is the sample size.

Confidence Interval = 50 ± (1.96 * (5 / √n))

Interpretation:
We are 95% confident that the true population mean lies within the interval [48.04, 51.96] based on the given sample data.

Q8: Margin of Error and Sample Size:
The margin of error (MOE) is the range around a sample estimate (like a mean or proportion) within which we expect the true population parameter to fall with a certain level of confidence. A larger sample size generally leads to a smaller margin of error.

Example:
Suppose you want to estimate the average age of voters in a city with a 95% confidence level. You can conduct a survey with a small sample size of 100 or a larger sample size of 1000. The MOE for each case would be smaller for the larger sample size, indicating a more precise estimate.

For instance:
- With a sample size of 100, the MOE might be ±3 years.
- With a sample size of 1000, the MOE might be ±1 year.

A larger sample size provides more information about the population, leading to a narrower range of uncertainty.

Q9: Z-Score Calculation and Interpretation:
Given a data point with a value of 75, a population mean (μ) of 70, and a population standard deviation (σ) of 5, the formula to calculate the z-score is:

Z = (x - μ) / σ

Z = (75 - 70) / 5
Z = 1

Interpretation:
The data point of 75 is 1 standard deviation above the mean.

- Null hypothesis (H₀): The drug is not significantly effective (μ = 0).

Compare the t-statistic to the critical t-value:
Since 13.42 > 2.0096, we reject the null hypothesis.

Conclusion:
Based on the t-test, there is sufficient evidence to conclude that the weight loss drug is significantly effective at a 95% confidence level.

Q11: Confidence Interval for Job Satisfaction Proportion:
Given:
- Sample proportion (p̂) = 0.65 (65%)
- Sample size (n) = 500
- Significance level (α) = 0.05 (for a 95% confidence level)

Calculate the standard error (SE) of the proportion:
SE = √(p̂ * (1 - p̂) / n)
SE = √(0.65 * 0.35 / 500)
SE ≈ 0.027

Find the critical z-value from a standard normal distribution table or calculator for a 95% confidence level. Let's assume the critical z-value is 1.96.

Calculate the margin of error (MOE):
MOE = z * SE
MOE = 1.96 * 0.027
MOE ≈ 0.053

Confidence Interval = p̂ ± MOE
Confidence Interval = 0.65 ± 0.053

Interpretation:
We are 95% confident that the true proportion of people satisfied with their job lies within the interval [0.597, 0.703] based on the survey data.
