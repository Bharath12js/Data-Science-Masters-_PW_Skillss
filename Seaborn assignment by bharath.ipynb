**Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

Solution:

Seaborn is a popular data visualization library in Python that is built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five commonly used plots that can be created using the Seaborn library:

1. Scatter Plot (`sns.scatterplot()`): A scatter plot is used to visualize the relationship between two continuous variables. Seaborn's scatter plot enhances the basic scatter plot with additional options for color encoding and grouping variables, making it useful for exploring patterns, correlations, and clusters in the data.

2. Bar Plot (`sns.barplot()`): A bar plot displays rectangular bars of varying heights to represent data. Seaborn's bar plot is often used to compare values across different categories or to show the distribution of a single variable. It can handle both categorical and numeric data and offers options for aggregating data and adding error bars.

3. Histogram (`sns.histplot()`): A histogram represents the distribution of a continuous variable by dividing the data into bins and displaying the frequency or count of values falling into each bin. Seaborn's histogram plot provides various customization options, such as choosing the number of bins, adding kernel density estimation (KDE) plots, and handling categorical data.

4. Heatmap (`sns.heatmap()`): A heatmap is a graphical representation of data where the values in a matrix are represented as colors. Seaborn's heatmap is particularly useful for visualizing correlation matrices or displaying the intensity of values across two dimensions. It provides options for customizing color palettes, adding annotations, and hierarchical clustering.

5. Line Plot (`sns.lineplot()`): A line plot displays data points connected by straight lines and is commonly used to show trends over time or any ordered sequence. Seaborn's line plot enhances the basic line plot with additional features, such as confidence intervals and flexible data aggregation capabilities.

Each of these plots in Seaborn offers various customization options, such as controlling colors, adding labels, handling categorical data, and enhancing visual aesthetics. They are commonly used for data exploration, statistical analysis, and presenting insights in a visually appealing manner.

Remember to import the Seaborn library (`import seaborn as sns`) before using these plot functions.

**Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.**

Solution:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot with different events and regions
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Set the title and labels
plt.title("FMRI Signal Over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Display the plot
plt.show()
```

**Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.**
    
Solution:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# First box plot: pclass vs. age
sns.boxplot(data=titanic, x='pclass', y='age')

# Set the title and labels
plt.title("Passenger Class vs. Age")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Display the plot
plt.show()

# Second box plot: pclass vs. fare
sns.boxplot(data=titanic, x='pclass', y='fare')

# Set the title and labels
plt.title("Passenger Class vs. Fare")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

# Display the plot
plt.show()
```

**Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.**

Solution:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Plot histogram with hue for 'cut' column
sns.histplot(data=diamonds, x='price', hue='cut')

# Set the title and labels
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")

# Display the plot
plt.show()
```

**Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.**

```python
import seaborn as sns

# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Plot pair plot with hue for 'species' column
sns.pairplot(data=iris, hue='species')

# Display the plot
plt.show()
```

**Que 6: Use the "flights" dataset from seaborn to plot a heatmap.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix form
flights_matrix = flights.pivot("month", "year", "passengers")

# Create a heatmap
sns.heatmap(flights_matrix, cmap="YlGnBu")

# Set the title and labels
plt.title("Flights Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")

# Display the plot
plt.show()
```
